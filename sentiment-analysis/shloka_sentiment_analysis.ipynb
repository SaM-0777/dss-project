# Shlokas Preprocessing (english)

```python
# lowercase tokenization remove special characters remove stop words and punctuation stemming
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string
string.punctuation

# import these modules
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def textpreprocess(txt):
    
    text=txt.lower()
    text=nltk.word_tokenize(text)
    newlist=[]
    
    for word in text:
        if word.isalnum():
            
            newlist.append(word)
            
    newlist2=[]        
            
    
    
    for word in newlist:
        
        if word not in stopwords.words('english') and word not in string.punctuation:
            newlist2.append(word)
            
     
    
    newlist3=[]
    for word in newlist2:
        
        newlist3.append(ps.stem(word))
        
    return " ".join(newlist3)    
    
    
            
            
            
```

```python
newdf
```

```python
textpreprocess(newdf["EngMeaning"][2])
```

```python
newdf["Transformed_engtext"]=newdf["EngMeaning"].apply(textpreprocess)
```

```python
from textblob import TextBlob
```

```python
# Define a function for sentiment analysis using TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
# Apply sentiment analysis to the DataFrame
newdf['sentiment'] = newdf["Transformed_engtext"].apply(get_sentiment)
```
